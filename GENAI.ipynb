```python
## NLP - Natural Language Processing -GEN AI used for text cleaning, analysis
### Lexical Analysis
text = "Hello World NLP is amazing! It helps the computer to understand the human language easily"
print(text)
```

```python
###Tokenisation - breaking the sentences into different words which are called as tokens
tokens = text.split()
print(tokens)
```

```python
### NLP - NLTK(Natural Language toolkit) - It is a library which gives NLP functionality
!pip install nltk

import nltk
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt_tab')
```

```python
sent_tokens = sent_tokenize(text)
print(sent_tokens)
```

```python
word_tokens = word_tokenize(text)
print("Actaul text :", text)
print(word_tokens)
```

```python
###case conversion #### text processing
lower_tokens = []
for i in word_tokens:
  k = i.lower()
  lower_tokens.append(k)
print(lower_tokens)
```

```python
lower_tokens  = [i.lower() for i in word_tokens]  ###it is faster and easy one line
print(lower_tokens)
```

```python
### removing puntuation
import string
print(string.punctuation)
```

```python
cleaned_words = []
for i in lower_tokens:
  if i not in string.punctuation:
    cleaned_words.append(i)
print(cleaned_words)
```

```python
###stopwords not useful and repetative remove those
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english') )
print(stop_words)
```

```python
cleaned_tokens= [i for i in cleaned_words if i not in stop_words]
cleaned_tokens
```

```python
from nltk.probability import FreqDist
freq = FreqDist(cleaned_tokens)
freq
```

```python
print(freq.most_common(4))
```
